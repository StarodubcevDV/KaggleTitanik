```
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegressionCV
import random
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```
test = pd.read_csv('test.csv')
train = pd.read_csv('train.csv')
gs = pd.read_csv('gender_submission.csv')
```

```
train
```

```
def cabines(x):
  if x == np.NaN:
    return np.NaN
  else:
    sum = 0
    dt = {'A': 7000, 'B': 6000, 'C': 5000, 'D': 4000, 'E': 3000, 'F': 2000, 'G': 1000}
    for i in str(x).split(' ')[0]:
      if i in dt.keys():
        sum += int(str(x).split(' ')[0].replace(i, '0')) + dt[i]
    return sum
```

```
train_cut = train.drop(labels=['Name', 'Ticket'], axis=1)
train_cut['Age'] = train_cut['Age'].fillna(random.randint(int(train_cut['Age'].min()), int(train_cut['Age'].max())))
train_cut['Sex'] = train_cut['Sex'].map({'male':0, 'female':1})
train_cut['Embarked'] = train_cut['Embarked'].map({'S':0, 'C':1, 'Q': 2})
train_cut['Cabin'] = train_cut['Cabin'].apply(lambda x: cabines(x))
train_cut['Cabin'] = train_cut['Cabin'].apply(lambda x: random.randint(1000, 1150) if x == 0 else x)
train_cut['Cabin'] /= 1000
train_cut = train_cut.fillna(0)

test_cut = test.drop(labels=['Name', 'Ticket'], axis=1)
test_cut['Age'] = test_cut['Age'].fillna(random.randint(int(test_cut['Age'].min()), int(test_cut['Age'].max())))
test_cut['Sex'] = test_cut['Sex'].map({'male':0, 'female':1})
test_cut['Embarked'] = test_cut['Embarked'].map({'S':0, 'C':1, 'Q': 2})
test_cut['Cabin'] = test_cut['Cabin'].apply(lambda x: cabines(x))
test_cut['Cabin'] = test_cut['Cabin'].apply(lambda x: random.randint(1000, 1150) if x == 0 else x)
test_cut['Cabin'] /= 1000
test_cut = test_cut.fillna(0)
```

```
X_train = train_cut.drop(labels=['Survived'], axis=1)
y_train = train_cut['Survived']
X_test = test_cut
```

```
lr = LogisticRegressionCV(cv=3, max_iter=3000)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
```
